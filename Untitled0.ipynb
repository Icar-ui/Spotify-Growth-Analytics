```python
# 1. ÿ™ÿ´ÿ®Ÿäÿ™ ÿßŸÑŸÖŸÉÿ™ÿ®ÿ© ÿßŸÑŸÑŸä ŸÉÿ™ÿ≠ŸÖŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÖŸÜ Kaggle
!pip install kagglehub

import kagglehub
import pandas as pd

# 2. ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÄ Dataset ÿØŸäÿ±ŸäŸÉÿ™
# ŸáÿßÿØ ÿßŸÑŸÉŸàÿØ ÿ∫ÿßÿØŸä Ÿäÿ¨Ÿäÿ® ÿßŸÑŸÖŸÑŸÅ ŸÖŸÜ Kaggle ŸàŸäÿ≠ÿ∑Ÿà ŸÅÿßŸÑŸÖŸÉŸäŸÜÿ© ÿØŸäÿßŸÑ Google
path = kagglehub.dataset_download("maharshipandya/-spotify-tracks-dataset")

# 3. ŸÇÿ±ÿßÿ°ÿ© ÿßŸÑŸÖŸÑŸÅ
# ÿßŸÑŸÉŸàÿØ ŸÉŸäÿ¥ŸàŸÅ ŸÅŸäŸÜ ŸÜÿ≤ŸÑ ÿßŸÑŸÖŸÑŸÅ ŸàŸÉÿßŸäŸÅÿ™ÿ≠Ÿà ÿ®ŸÄ Pandas
import os
csv_path = os.path.join(path, "dataset.csv")
df = pd.read_csv(csv_path)

# 4. ÿ¨ÿ±ÿ® Ÿàÿßÿ¥ ÿÆÿØÿßŸÖ (ÿ∫ÿßŸäŸàÿ±ŸäŸÉ ÿ£ŸàŸÑ 5 ÿ≥ÿ∑Ÿàÿ±)
print("‚úÖ ÿßŸÑÿ™ŸÖÿ±ŸäŸÜÿ© ÿØÿßÿ≤ÿ™ ÿ®ŸÜÿ¨ÿßÿ≠! ŸáÿßŸáŸä ÿ£ŸàŸÑ 5 ÿ≥ÿ∑Ÿàÿ± ŸÖŸÜ ÿ®ŸäÿßŸÜÿßÿ™ Spotify:")
df.head()
```

```python
# ŸÉŸàÿØ ÿ®ÿßÿ¥ ŸÜÿπÿ±ŸÅŸà ÿ£ŸÉÿ´ÿ± Genres ŸÖÿ¥ŸáŸàÿ±ŸäŸÜ
top_genres = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)

print("üî• Top 10 Most Popular Genres on Spotify:")
print(top_genres)

# ÿ±ÿ≥ŸÖ ŸÖÿ®ŸäÿßŸÜŸä ÿ®ÿßÿ¥ ÿ™ÿ®ÿßŸÜ ÿßŸÑÿÆÿØŸÖÿ© ŸÜŸÇŸäÿ©
top_genres.plot(kind='bar', color='green', figsize=(10,5))
plt.title('Top 10 Genres by Popularity')
plt.ylabel('Average Popularity Score')
plt.show()
```
