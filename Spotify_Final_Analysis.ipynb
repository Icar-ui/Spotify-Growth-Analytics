```python
# 1. Install necessary libraries
!pip install kagglehub

import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import os

# 2. Download the Spotify Dataset
# Using kagglehub to fetch the latest version automatically
path = kagglehub.dataset_download("maharshipandya/-spotify-tracks-dataset")

# 3. Load the data
csv_path = os.path.join(path, "dataset.csv")
df = pd.read_csv(csv_path)

# 4. Marketing Analysis: Identify Top 10 Genres by Popularity
# This provides insights into which music categories attract the most users
top_genres = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)

print("✅ Analysis Successful! Results below:")
print(top_genres)

# 5. Professional Visualization
# Using Spotify's official green hex code: #1DB954
plt.figure(figsize=(12,6))
top_genres.plot(kind='bar', color='#1DB954')
plt.title('Spotify Growth Strategy: Top 10 Most Popular Genres', fontsize=14)
plt.xlabel('Music Genre', fontsize=12)
plt.ylabel('Average Popularity Score', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
